```
import time
import matplotlib.pyplot as plt
import sys

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n_values = list(range(1, 30))
execution_times = []
space_usage = []

for n in n_values:
    start = time.time()
    fibonacci(n)
    end = time.time()
    execution_times.append(end - start)
    frame_size = sys.getsizeof(n)
    space_usage.append(n * frame_size)

plt.figure(figsize=(8, 5))
plt.plot(n_values, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(n_values, space_usage, marker='o', color='green', label='Space (bytes)')
plt.title('Time vs Space Complexity of Recursive Fibonacci')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```
import time
import matplotlib.pyplot as plt
import sys

def fibonacci_dp(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib

n_values = list(range(1, 1001, 50))
time_dp = []
space_dp = []

for n in n_values:
    start = time.time()
    fibonacci_dp(n)
    end = time.time()
    time_dp.append(end - start)
    fib_list = fibonacci_dp(n)
    space_dp.append(sys.getsizeof(fib_list))

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(n_values, time_dp, marker='o', color='green', label='Time')
ax1.set_xlabel('n')
ax1.set_ylabel('Execution Time (seconds)', color='green')
ax1.tick_params(axis='y', labelcolor='green')

ax2 = ax1.twinx()
ax2.plot(n_values, space_dp, marker='x', color='orange', linestyle='--', label='Space')
ax2.set_ylabel('Space (bytes)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Time and Space Complexity of DP Fibonacci')
plt.grid(True)
plt.show()
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n-1):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
time_taken = []
space_used = []

for size in sizes:
    arr = [random.randint(1, 10000) for _ in range(size)]
    space_used.append(sys.getsizeof(arr))
    start = time.time()
    bubble_sort(arr)
    end = time.time()
    time_taken.append(end - start)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(sizes, time_taken, marker='o', color='green', label='Time')
ax1.set_xlabel('Array Size (n)')
ax1.set_ylabel('Execution Time (seconds)', color='green')
ax1.tick_params(axis='y', labelcolor='green')

ax2 = ax1.twinx()
ax2.plot(sizes, space_used, marker='x', color='orange', linestyle='--', label='Space')
ax2.set_ylabel('Space (bytes)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Time and Space Complexity of Bubble Sort')
plt.grid(True)
plt.show()
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
    return arr

sizes = [100, 500, 1000, 2000, 3000]
time_taken = []
space_used = []

for size in sizes:
    arr1 = [random.randint(1, 10000) for _ in range(size)]
    arr2 = [random.randint(1, 10000) for _ in range(size)]
    merged_arr = arr1 + arr2

    space_used.append(sys.getsizeof(merged_arr))
    start = time.time()
    merge_sort(merged_arr)
    end = time.time()
    time_taken.append(end - start)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(sizes, time_taken, marker='o', color='green', label='Time')
ax1.set_xlabel('Size of Each Array (n)')
ax1.set_ylabel('Execution Time (seconds)', color='green')
ax1.tick_params(axis='y', labelcolor='green')

ax2 = ax1.twinx()
ax2.plot(sizes, space_used, marker='x', color='orange', linestyle='--', label='Space')
ax2.set_ylabel('Space (bytes)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Time and Space Complexity of Merge Sort (Two Arrays Merged)')
plt.grid(True)
plt.show()
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1

sizes = [100, 500, 1000, 5000, 10000, 20000, 30000]
time_taken = []
space_used = []

for size in sizes:
    arr = sorted([random.randint(1, 100000) for _ in range(size)])
    x = arr[random.randint(0, size - 1)]

    space_used.append(sys.getsizeof(arr))
    start = time.time()
    binary_search(arr, x)
    end = time.time()
    time_taken.append(end - start)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(sizes, time_taken, marker='o', color='blue', label='Time')
ax1.set_xlabel('Array Size (n)')
ax1.set_ylabel('Execution Time (seconds)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(sizes, space_used, marker='x', color='orange', linestyle='--', label='Space')
ax2.set_ylabel('Space (bytes)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Time and Space Complexity of Binary Search')
plt.grid(True)
plt.show()

```

```
import time
import matplotlib.pyplot as plt
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n-1):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]
        merge_sort(left)
        merge_sort(right)
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
    return arr

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

sizes = [100, 500, 1000, 2000]
bubble_times = []
selection_times = []
insertion_times = []
merge_times = []
quick_times = []

for size in sizes:
    arr = [random.randint(1, 10000) for _ in range(size)]
    start = time.time()
    bubble_sort(arr.copy())
    bubble_times.append(time.time() - start)
    start = time.time()
    selection_sort(arr.copy())
    selection_times.append(time.time() - start)
    start = time.time()
    insertion_sort(arr.copy())
    insertion_times.append(time.time() - start)
    start = time.time()
    merge_sort(arr.copy())
    merge_times.append(time.time() - start)
    start = time.time()
    quick_sort(arr.copy())
    quick_times.append(time.time() - start)

plt.figure(figsize=(10, 6))
plt.plot(sizes, bubble_times, marker='o', label='Bubble Sort')
plt.plot(sizes, selection_times, marker='s', label='Selection Sort')
plt.plot(sizes, insertion_times, marker='^', label='Insertion Sort')
plt.plot(sizes, merge_times, marker='d', label='Merge Sort')
plt.plot(sizes, quick_times, marker='x', label='Quick Sort')
plt.xlabel('Array Size (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Time Complexity Comparison of Sorting Algorithms')
plt.legend()
plt.grid(True)
plt.show()
```
